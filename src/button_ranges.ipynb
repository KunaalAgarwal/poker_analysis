## RFI (Raise first in, Open) range

Only options from RFI are Raise/Fold. No open limping in my strategy.

Playing a "merged" strategy. Rather than having a polarized range with a section of extremely strong value hands and extremely weak bluffs, I include bluffs as playable hands that are slightly weaker than my strongest hands. 

Range percent approximately = mean hand percentile indicates that a merged range has been used. 

A or K high percentage vitally important to be high. Targeting around 55%. 
Given the breakeven for a bluff as pot odds being laid after the bet and assuming that roughly 45% of our range is a bluff on A/K high boards. We can c-bet whole range for Half-pot profitably.

# 3! Response range

Determination of 3bet call range size via MDF. Assuming a 3bet sizing of around 3.5x (3x + dead money usually), alpha = (3.5 / 3.5 + 1 + dead money), MDF = 1 - (3.5 / 4.5 + dead money). MDF is approximately 30%. Thus we need to call with at minimum 30% of our BTN open range. 

Also consider value to bluff ratios. Pot odds with a 3.5x bet are: 
2.5 / ((1 + 3.5 + dead money) + 2.5) = approximately 25%. People are often 3betting bluffing far more than 25% of the time however, but will also overcall 4bets (with marginal hands) and thus the best strategy is to play a polarized 4bet strategy and call 3bets at slightly higher than MDF. 

4bet range should be around 60% value, 40% bluffs. Have 36 value and 23 bluff combos.
Leads to a 2.5x 4bet size from IP.
